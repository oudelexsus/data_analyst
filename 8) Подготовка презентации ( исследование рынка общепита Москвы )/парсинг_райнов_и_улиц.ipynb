```python
import requests as re
from bs4 import BeautifulSoup


url = 'http://mosopen.ru/streets'
r = re.get(url)
soup = BeautifulSoup(r.text, 'lxml')

spisok_strok_href_streets = soup.find('table', class_ = 'regions_list').findAll('a')

areas_names = []
areas_link = []

for stroka in spisok_strok_href_streets:
    areas_names.append(stroka.text)
    areas_link.append(stroka.get('href'))

areas_link
```

Добавим в наш большой цикл эти дополнения

```python
street_names = []

for url in areas_link:
            
            street_names_stroka = []
            r = re.get(url)
            soup = BeautifulSoup(r.text, 'lxml')
            try: 
                spisok_strok_streets_in_area = soup.find('div', class_ = 'double_block clearfix').findAll('a')
            except AttributeError:
                  browser.get(url)
                  spisok_streets = browser.find_element(By.XPATH, '//*[@id="content"]/ul[2]').text.split('\n')
                  street_names.append(spisok_streets)
                  continue

            for stroka in spisok_strok_streets_in_area:
                street_names_stroka.append(stroka.text)

            street_names.append(street_names_stroka)


street_names
```

```python
import pandas as pd



new_street_names = []

for spisok in street_names:
    new_spisok = []
    for street in spisok:
        street = street.replace(',', '')
        new_spisok.append(street)
    
    new_street_names.append(new_spisok)

new_street_names
```

```python
import numpy as np

data = pd.DataFrame()
for i in np.arange(0, len(areas_names)):
    slovar = [{areas_names[i]: j} for j in new_street_names[i]]
    data1 = pd.DataFrame(slovar).rename({areas_names[i]: 'Название улицы/поселения'}, axis = 1)
    data1['Район'] = areas_names[i]
    data = pd.concat([data, data1], axis=0)

data
```
